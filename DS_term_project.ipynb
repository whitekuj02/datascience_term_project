```python
# 각 features에 대한 연관성
# plot the corrleation matrix as a heatmap
# seaborn의 heatmap() 함수 사용해서 상관관계 히트맵 그림
# 그 중 stroke 부분 중점
# age(0~20, 20~40, 40~60, 60~85), hypertension(0,1), heart_disease(0,1), ever_married(Yes, No), work_type(Private, children, Self-employed, Govt_job),  
# Residence_type(Urban, Rural), avg_glucose_level(55~100, 100~200, 200~275), bmi(10~30, 30~50, 50~70, 70~), smoking_status(formerly smoked, never smoked, Unknown, smokes). stroke(0,1)

# read data
daFt = pd.read_csv("healthcare-dataset-stroke-data.csv")

# age(0~20, 20~40, 40~60, 60~85)
# 'age' 카테고리의 값을 더미 변수(dummy variable)로 변환
# Create dummy variable for age
age_dummies = pd.get_dummies(daFt['age'], prefix='age')
# concat() 함수를 사용하여 더미 변수를 기존 데이터프레임에 병합
# concentenate dummy variable with the original dataframe
daFt = pd.concat([daFt, age_dummies], axis=1)
daFt['age_0_20']=daFt['age'][(daFt['age']>=0) & (daFt['age']<20)].dropna()
daFt['age_20_40']=daFt['age'][(daFt['age']>=20) & (daFt['age']<40)].dropna()
daFt['age_40_60']=daFt['age'][(daFt['age']>=40) & (daFt['age']<60)].dropna()
daFt['age_60_85']=daFt['age'][(daFt['age']>=60) & (daFt['age']<85)].dropna()

# ever_married(Yes, No)
# 'ever_married' 카테고리의 값을 더미 변수(dummy variable)로 변환
# Create dummy variable for ever_married
marry_dummies = pd.get_dummies(daFt['ever_married'], prefix='ever_married')
# concat() 함수를 사용하여 더미 변수를 기존 데이터프레임에 병합
# concentenate dummy variable with the original dataframe
daFt = pd.concat([daFt, marry_dummies], axis=1)

# work_type(Private, children, Self-employed, Govt_job)
# 'work_type' 카테고리의 값을 더미 변수(dummy variable)로 변환
# Create dummy variable for work_type
work_dummies = pd.get_dummies(daFt['work_type'], prefix='work_type')
# concat() 함수를 사용하여 더미 변수를 기존 데이터프레임에 병합
# concentenate dummy variable with the original dataframe
daFt = pd.concat([daFt, work_dummies], axis=1)

# Residence_type(Urban, Rural)
# 'Residence_type' 카테고리의 값을 더미 변수(dummy variable)로 변환
# Create dummy variable for Residence_type
resid_dummies = pd.get_dummies(daFt['Residence_type'], prefix='Residence_type')
# concat() 함수를 사용하여 더미 변수를 기존 데이터프레임에 병합
# concentenate dummy variable with the original dataframe
daFt = pd.concat([daFt, resid_dummies], axis=1)

# avg_glucose_level(55~100, 100~200, 200~275)
# 'avg_glucose_level' 카테고리의 값을 더미 변수(dummy variable)로 변환
# Create dummy variable for age
glu_dummies = pd.get_dummies(daFt['avg_glucose_level'], prefix='avg_glucose_level')
# concat() 함수를 사용하여 더미 변수를 기존 데이터프레임에 병합
# concentenate dummy variable with the original dataframe
daFt = pd.concat([daFt, glu_dummies], axis=1)
daFt['glu_55_100']=daFt['avg_glucose_level'][(daFt['avg_glucose_level']>=55) & (daFt['avg_glucose_level']<100)].dropna()
daFt['glu_100_200']=daFt['avg_glucose_level'][(daFt['avg_glucose_level']>=100) & (daFt['avg_glucose_level']<200)].dropna()
daFt['glu_200_275']=daFt['avg_glucose_level'][(daFt['avg_glucose_level']>=200) & (daFt['avg_glucose_level']<275)].dropna()

# bmi(10~30, 30~50, 50~70, 70~)
# 'bmi' 카테고리의 값을 더미 변수(dummy variable)로 변환
# Create dummy variable for bmi
bmi_dummies = pd.get_dummies(daFt['bmi'], prefix='bmi')
# concat() 함수를 사용하여 더미 변수를 기존 데이터프레임에 병합
# concentenate dummy variable with the original dataframe
daFt = pd.concat([daFt, bmi_dummies], axis=1)
daFt['bmi_10_30']=daFt['bmi'][(daFt['bmi']>=10) & (daFt['bmi']<30)].dropna()
daFt['bmi_30_50']=daFt['bmi'][(daFt['bmi']>=30) & (daFt['bmi']<50)].dropna()
daFt['bmi_50_70']=daFt['bmi'][(daFt['bmi']>=50) & (daFt['bmi']<70)].dropna()
daFt['bmi_70']=daFt['bmi'][(daFt['bmi']>=70)].dropna()

# smoking_status(formerly smoked, never smoked, Unknown, smokes)
# 'smoking_status' 카테고리의 값을 더미 변수(dummy variable)로 변환
# Create dummy variables for 'smoking_status'
smoking_dummies = pd.get_dummies(daFt['smoking_status'], prefix='smoking')
# concat() 함수를 사용하여 더미 변수를 기존 데이터프레임에 병합
# concentenate dummy variable with the original dataframe
daFt = pd.concat([daFt, smoking_dummies], axis=1)

corr = daFt[['stroke','age_0_20', 'age_20_40', 'age_40_60', 'age_60_85', 'hypertension', 'heart_disease', 'ever_married_Yes', 'ever_married_No',
             'work_type_Private', 'work_type_children', 'work_type_Self-employed', 'work_type_Govt_job',
             'Residence_type_Urban', 'Residence_type_Rural', 'glu_55_100', 'glu_100_200', 'glu_200_275',
             'bmi_10_30', 'bmi_30_50', 'bmi_50_70', 'bmi_70', 'smoking_never smoked', 'smoking_formerly smoked', 'smoking_smokes', 'smoking_Unknown']].corr()

# stroke와의 상관관계만을 보기 위해 맨 위의 행만 출력하게 함
corr_top_row = corr.head(1)
plt.figure(figsize=(20, 1))  # Set figure size
sns.heatmap(corr_top_row, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of all features")
plt.show()

# the highest correlation Top 5
top_corr_features = corr['stroke'].sort_values(ascending=False)[1:6]
# the lowest corrleation Top 5
bottom_corr_features = corr['stroke'].sort_values()[0:5]

print("Top 5 features with highest correlation to stroke:")
print(top_corr_features)
print("\nBottom 5 features with lowest correlation to stroke:")
print(bottom_corr_features)

# 이 결과를 토대로 stroke과 가장 연관성 있는 것들 5개를 꼽으면 (치수는 소수점 셋째자리에서 반올림)
# age_60_85 (0.17), age_40_60 (0.14), glu_100_200 (0.14), heart_disease (0.13), hyertension (0.13) 가 있고
# 가장 연관성 없는 것들 5개를 꼽으면 (치수는 소수점 셋째자리에서 반올림)
# ever_married_No (-0.11), work_type_children (-0.084), smoking_Unknown (-0.056), glu_55_100 (-0.019), age_0_20 (-0.015) 이 있다

# 즉, 나이가 높을 수록 stroke가 발생할 확률이 높으며 나이가 낮을 수록 발병 확률이 낮다는 것을 알 수 있다.
# 또한 포도당 수치가 높으면 발병 확률이 높고, 낮으면 상대적으로 발병 확률이 낮다는 것을 알 수 있다.
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
#Scale numerical features in independent variables using a various scaling method
def scaling_data(X, numeric_col_names, method='std'):   
    if method == 'std':      
        std = StandardScaler()     
        return std.fit_transform(X[numeric_col_names])   
    elif method == 'minmax':       
        min_max = MinMaxScaler()   
        return min_max.fit_transform(X[numeric_col_names])  
    elif method == 'robust':       
        robust = RobustScaler()      
        return robust.fit_transform(X[numeric_col_names])  
    else:        
        return -1
```

```python
# train_test_split()
# k_fold test train 랜덤으로 몇번 뽑아서 학습 (리나)
# 이 부분은 위에서의 data analysis에 사용한 model을 사용하여 새로운 model을 효율적으로 학습시켜주세요
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix

def kFordEv(n, model):    
    kfold = KFold(n_splits=n)   
    cv_accuracy = []    
    n_iter = 0   
    for train_index, test_index in kfold.split(X):        
        x_train, x_test = X.iloc[train_index], X.iloc[test_index]    
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]       
        # training       
        model.fit(x_train, y_train)        
        # predict        
        pred = model.predict(x_test)      
        n_iter += 1       
        # Measure accuracy every iteration(Round to 5 decimal places)     
        accuracy = np.round(accuracy_score(y_test, pred), 5)       
        train_size = x_train.shape[0]       
        test_size = x_test.shape[0]      
        print('\n>{0}th cross-validation accuracy : {1},  training data size : {2},  validation data size : {3}'             
              .format(n_iter, accuracy, train_size, test_size))      
        print('>{0} validation set index : {1}'.format(n_iter, test_index))        
        cv_accuracy.append(accuracy)   
    return np.mean(cv_accuracy)

'''
KFold result of Base of Decision Tree
'''
print('\n# result 5-fold :')
kfold_mean_accuracy = kFordEv(5, dt_clf)
print('->> average validation accuracy :', kfold_mean_accuracy)
```
